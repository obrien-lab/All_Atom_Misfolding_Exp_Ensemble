# Aims of this code:
    1. Read experiment excel file for cutsites
    2. extract only fragment from ISPE
    3. Extract cutsite that has fragment with |R/N| > 2 and significant at 0.05 pvalue level after FDR correction

```python
import pandas as pd
```

```python
df = pd.read_excel('EXP/N_v_R_cut_site.xlsx')
```

```python
ispe = df[df['Gene']=='ispE']
```

```python
ispe
```

```python
# Significant cutsite: -log10(adj. pvalue) > 1.301 <=> adj.pvalue < 0.05
# Log2 FC > 1 <=> abs(R/N) > 2
# Must be half-tryptic
ispe_significant = ispe[(abs(ispe['Log2 FC'])>=1) & (ispe['-Log10 Adj. P-value']>=1.301) & (ispe['Half Tryptic']==True)]
```

```python
ispe_significant.reset_index()
```

```python
# Extract the numeric part and convert it to an integer
ispe_significant['Numeric Part'] = ispe_significant['Cut Site'].str.extract('(\d+)').astype(int)
ispe_significant
```

```python

# Sort the DataFrame by the numeric part
ispe_significant_sorted = ispe_significant.sort_values(by='Numeric Part')

# Drop the temporary 'Numeric Part' column if not needed
ispe_significant_sorted = ispe_significant_sorted.drop(columns=['Numeric Part'])

print(ispe_significant_sorted)
```

```python
ispe_significant_sorted_reset = ispe_significant_sorted.reset_index()
```

```python
ispe_significant_sorted_reset.to_excel('ispe_lipms_significant_cut_site.xlsx')
```
