```python
import numpy as np
import time
```

```python
def get_ci_95(data):
    nboot = 100000
    mean_vals = np.zeros(nboot)

    for i in range(nboot):
        idx = np.random.choice(data, size=len(data))
        mean_vals[i] = np.mean(idx)

    mean = np.mean(data)
    lower_bound = np.percentile(mean_vals, 2.5)
    upper_bound = np.percentile(mean_vals, 97.5)

    return [mean, lower_bound, upper_bound]
```

```python
cutsite_range = {
    "R69": [64,75],
    "M75": [70,81],
    "A78": [73,84],
    "I159": [154,165],
    "L202": [197,208],
    "K204": [199,210],
    "E206": [201,212],
    "F207": [202,213],
    "A254": [249,260],
    "R255": [250,261],
    "L258": [253,264],
    "Q260": [255,266],
    "A261": [256,267],
}
```

```python
for key, value in cutsite_range.items():    
    start_time = time.time()
    for st in range(6):
        dat = np.load(f"sim/ASA_cutsite_then_state/ASA_{key}_state_{st}.npy")
        print(f"{key}    state: {st}")
        results = get_ci_95(dat)
        print(f"{results[0]:.3f}\t[{results[1]:.3f}, {results[2]:.3f}]")

    end_time = time.time()
    
    print("Execution time:", end_time - start_time, "seconds")
    print("**************************************")
```
