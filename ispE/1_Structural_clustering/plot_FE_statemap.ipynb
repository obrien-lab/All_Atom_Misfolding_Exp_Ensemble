```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pyemma
import mdtraj
from pyemma.util.contexts import settings
import networkx as nx
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Matplotlib setting
fontsize = 12 # set variable for fontsize
linewidth = 2 # set variable for line width
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.sans-serif'] = ['Arial']
matplotlib.rcParams['axes.labelsize'] = 'small'
matplotlib.rcParams['axes.linewidth'] = 1
matplotlib.rcParams['lines.markersize'] = 4
matplotlib.rcParams['xtick.major.width'] = 1
matplotlib.rcParams['ytick.major.width'] = 1
matplotlib.rcParams['xtick.labelsize'] = 'x-small'
matplotlib.rcParams['ytick.labelsize'] = 'x-small'
matplotlib.rcParams['legend.fontsize'] = 'x-small'
matplotlib.rcParams['figure.dpi'] = 600
```

```python
msm_data=np.load('msm_data.npz', allow_pickle=True)
```

```python
msm_data.files
```

```python
data = msm_data['raw_Q_G']
```

```python
conc_data=np.concatenate(data)
```

```python
meta_dtrajs = msm_data['meta_dtrajs']
```

```python
fig1, ax1 = plt.subplots()
# pyemma.plots.plot_free_energy(data_refold_conc[:,0], data_refold_conc[:,1], cbar_label='-ln(P)', ax=ax1)
# pyemma.plots.plot_free_energy(conc_data[:,0], conc_data[:,1])
pyemma.plots.plot_free_energy(conc_data[:,0], conc_data[:,1], cbar_label='-ln(P)/kT', ax=ax1)
plt.xlim([0.65,1])
plt.xlabel('Q', fontsize=fontsize+2)
plt.ylabel('G', fontsize=fontsize+2)
# Remove the top-right border
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_linewidth(linewidth) # make left axis thicker
ax1.spines['bottom'].set_linewidth(linewidth) # make bottom axis thicker
ax1.xaxis.set_tick_params(width=linewidth) # make x-axis tick marks thicker
ax1.yaxis.set_tick_params(width=linewidth) # make y-axis tick marks thicker
fig1.savefig(f'ispE_FE.png', dpi=1000)
```

```python
meta_dtraj_flatten = np.array(meta_dtrajs.flatten(), dtype=int)
```

```python
fig2, ax2 = plt.subplots()
# pyemma.plots.plot_state_map(conc_data[:,0], conc_data[:,1], meta_dtraj_flatten)
pyemma.plots.plot_state_map(conc_data[:,0], conc_data[:,1], meta_dtraj_flatten, ax=ax2)
plt.xlim([0.65,1])
plt.xlabel('Q', fontsize=fontsize+2)
plt.ylabel('G', fontsize=fontsize+2)
# Remove the top-right border
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_linewidth(linewidth) # make left axis thicker
ax2.spines['bottom'].set_linewidth(linewidth) # make bottom axis thicker
ax2.xaxis.set_tick_params(width=linewidth) # make x-axis tick marks thicker
ax2.yaxis.set_tick_params(width=linewidth) # make y-axis tick marks thicker
fig2.savefig(f'ispE_Statemap.png', dpi=1000)
```
